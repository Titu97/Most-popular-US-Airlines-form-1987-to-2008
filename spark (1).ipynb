```python
t  = spark.read.csv("gs://dataproc-staging-europe-west4-1041851436623-ceaakthz/y",header=True, inferSchema=True)
```

```python
import pandas
from collections import Counter

for i in range(2):
    ty= t.filter(t.Year == 1987+i)    
    ty1=ty.groupby('UniqueCarrier').count() 
    a=ty1.sort("count", ascending=False).take(10)
    
    we=[]
    for k in range(10):
        s=[a[k][0]]*a[k][1]
        we.extend(s)
    
    letter_counts = Counter(we)
    df = pandas.DataFrame.from_dict(letter_counts, orient='index')
    df.plot(title=str(1987+i),kind='bar')
 
```
